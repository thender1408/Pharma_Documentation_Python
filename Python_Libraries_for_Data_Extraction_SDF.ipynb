## Text and Table Extraction with pdfplumber

pdfplumber excels at extracting tables and provides more control over text extraction.

Open a PDF:

```python
import pdfplumber

# Load the PDF
with pdfplumber.open("sample-sdf-document - Cytiva.pdf") as pdf:
    # Get the total number of pages
    print(f"The PDF has {len(pdf.pages)} pages.")
```

Extract Text from a Page

```python
# Extract text from the first page
with pdfplumber.open("sample-sdf-document - Cytiva.pdf") as pdf:
    page = pdf.pages[0]
    print(page.extract_text())
```

### Extract Tables

  pdfplumber can automatically detect and extract tables from PDFs.

  **Important:** This works best with tables that have visible borders (lines). Many pharmaceutical documents use borderless tables which won't be detected automatically.

  In our sample:
  - **Page 1-2**: Real documents with borderless tables (won't extract)
  - **Page 3**: Certificate with bordered table (will extract)

```python
# Extract tables from different pages
# Try different pages to find tables in your SDF
with pdfplumber.open("sample-sdf-document - Cytiva.pdf") as pdf:
    for page_num, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        if tables:
            print(f"\n--- Tables found on Page {page_num + 1} ---")
            for table in tables:
                for row in table:
                    print(row)
```

### What About Borderless Tables?

  Pages 1 and 2 have tables that pdfplumber couldn't detect. For these, you'll need to:

  1. Extract the raw text
  2. Use regex patterns to find and parse the data

  You'll learn this technique in **Step-02** when we cover regular expressions.


```python
# Preview: Here's what the borderless table text looks like
# You'll learn to parse this with regex in Step-02

with pdfplumber.open("sample-sdf-document - Cytiva.pdf") as pdf:
    page = pdf.pages[0]  # Page 1 has a borderless table
    text = page.extract_text()

    # Find the table section
    lines = text.split('\n')
    for i, line in enumerate(lines):
        if 'Description' in line or 'High Flow' in line:
            print(line)
```

## Using PyMuPDF

PyMuPDF (also called fitz) is the most powerful library for extracting both text AND bounding box information. This is essential for understanding document layout.

Open a PDF and View Metadata:

```python
import fitz  # PyMuPDF

# Load the PDF
doc = fitz.open("sample-sdf-document.pdf")

# Print metadata
print("PDF Metadata:", doc.metadata)
print(f"\nTotal pages: {len(doc)}")
```

```python
# Extract text from the first page
page = doc[0]
print(page.get_text())
```

### Extract Text with Bounding Boxes

This is what makes PyMuPDF special - you can get the exact position of each text element on the page. This helps AI models understand document structure.

```python
text_data = page.get_text("dict")  # Extract text as a structured dictionary

for block in text_data.get("blocks", []):  # Loop through all blocks in the page

    # Check if 'lines' exists in the block
    if "lines" in block:
        for line in block["lines"]:
            for span in line.get("spans", []):  # Ensure spans exist
                bbox = span["bbox"]
                text = span["text"]
                print(f"Text: {text}, Bounding Box: {bbox}")
```

### Render Page as Image

You can also render PDF pages as images for visual inspection or further processing.

```python
# Render the first page as an image
pix = page.get_pixmap()
pix.save("page1.png")
print("Page saved as page1.png")
```

### Extract Text from All Pages

Loop through all pages to extract text from the entire document.

```python
# Loop through all pages and extract text
for page_num in range(len(doc)):
    page = doc[page_num]
    text = page.get_text("text")
    print(f"\n--- Page {page_num + 1} ---")
    print(text[:500])  # Print first 500 chars of each page
    print("...")
```

## Summary: Choosing the Right Library

| Library | Best For | Bounding Boxes |
|---------|----------|----------------|
| PyPDF2 | Quick text extraction | No |
| pdfplumber | Table extraction | Limited |
| PyMuPDF | Full extraction + layout | Yes |

For pharmaceutical SDF processing where understanding document layout matters, **PyMuPDF** is usually the best choice.
